```sql
USE QueensClassSchedule

SELECT 
    FirstName,
    LastName,
    COUNT(DISTINCT DepartmentName) AS NumberOfDepartments
FROM 
    [Departmental].[Instructor]
GROUP BY 
    FirstName,
    LastName
HAVING 
    COUNT(DISTINCT DepartmentName) > 1;
```

<span style="color: rgb(13, 13, 13); font-family: Söhne, ui-sans-serif, system-ui, -apple-system, &quot;Segoe UI&quot;, Roboto, Ubuntu, Cantarell, &quot;Noto Sans&quot;, sans-serif, &quot;Helvetica Neue&quot;, Arial, &quot;Apple Color Emoji&quot;, &quot;Segoe UI Emoji&quot;, &quot;Segoe UI Symbol&quot;, &quot;Noto Color Emoji&quot;; font-size: 16px; white-space-collapse: preserve; background-color: rgb(255, 255, 255);"><b>Proposition 2</b>: Find the total number of instructors in each department.</span>

```sql
USE QueensClassSchedule

SELECT 
    DepartmentName,
    COUNT(DISTINCT FirstName + ' ' + LastName) AS NumberOfInstructors
FROM 
    [Departmental].[Instructor]
GROUP BY 
    DepartmentName
ORDER BY 
    DepartmentName;
```

<span style="color: rgb(13, 13, 13); font-family: Söhne, ui-sans-serif, system-ui, -apple-system, &quot;Segoe UI&quot;, Roboto, Ubuntu, Cantarell, &quot;Noto Sans&quot;, sans-serif, &quot;Helvetica Neue&quot;, Arial, &quot;Apple Color Emoji&quot;, &quot;Segoe UI Emoji&quot;, &quot;Segoe UI Symbol&quot;, &quot;Noto Color Emoji&quot;; font-size: 16px; white-space-collapse: preserve; background-color: rgb(255, 255, 255);"><b>Proposition 3</b>: Find the classes taught by course along with the total enrollment, total class limit, and the percentage of enrollment.</span>

```sql
USE QueensClassSchedule

SELECT 
    CourseAbbreviation,
    COUNT(ClassID) AS NumberOfClasses,
    SUM(CAST(Enrolled AS INT)) AS TotalEnrollment,
    SUM(CAST(Limit AS INT)) AS TotalClassLimit,
    CASE 
        WHEN SUM(CAST(Limit AS INT)) > 0 
        THEN CONVERT(DECIMAL(5, 2), (SUM(CAST(Enrolled AS INT)) * 100.0) / SUM(CAST(Limit AS INT)))
        ELSE 0 
    END AS PercentageOfEnrollment
FROM 
    [CollegeClasses].[Class]
GROUP BY 
    CourseAbbreviation
ORDER BY 
    CourseAbbreviation;
```

<span style="color: rgb(13, 13, 13); font-family: Söhne, ui-sans-serif, system-ui, -apple-system, &quot;Segoe UI&quot;, Roboto, Ubuntu, Cantarell, &quot;Noto Sans&quot;, sans-serif, &quot;Helvetica Neue&quot;, Arial, &quot;Apple Color Emoji&quot;, &quot;Segoe UI Emoji&quot;, &quot;Segoe UI Symbol&quot;, &quot;Noto Color Emoji&quot;; font-size: 16px; white-space-collapse: preserve; background-color: rgb(255, 255, 255);"><b>Proposition 4</b>: Determine the number of in-person, hybrid, and web-enhanced classes each distinct department offers.</span>

```sql
USE QueensClassSchedule

SELECT 
    d.DepartmentName,
    SUM(CASE WHEN c.ModeOfInstruction = 'In-Person' THEN 1 ELSE 0 END) AS InPersonClasses,
    SUM(CASE WHEN c.ModeOfInstruction = 'Hybrid' THEN 1 ELSE 0 END) AS HybridClasses,
    SUM(CASE WHEN c.ModeOfInstruction = 'Web-Enhanced' THEN 1 ELSE 0 END) AS WebEnhancedClasses
FROM 
    [Departmental].[Department] d
JOIN 
    [CollegeClasses].[Class] c ON d.DepartmentID = c.CourseID -- Assuming CourseID in Class table maps to DepartmentID in Department table
GROUP BY 
    d.DepartmentName
ORDER BY 
    d.DepartmentName;
```

```sql
USE QueensClassSchedule

SELECT 
    BuildingID,
    BuildingName,
    RoomID,
    RoomNumber,
    COUNT(ClassID) AS NumberOfClasses
FROM 
    [CollegeClasses].[Class]
GROUP BY 
    BuildingID,
    BuildingName,
    RoomID,
    RoomNumber
HAVING 
    COUNT(ClassID) > 1
ORDER BY 
    BuildingName,
    RoomNumber;
```

**Proposition 6**: Find all courses that provide 4 credits.

```sql
USE QueensClassSchedule

SELECT 
    CourseAbbreviation,
    CourseName
FROM 
    [CollegeClasses].[Course]
WHERE 
    CourseCredit = 4
ORDER BY 
    CourseAbbreviation, 
    CourseName;
```

**Proposition 7**: Determine the number of rooms used in each building location.

```sql
USE QueensClassSchedule

SELECT 
    bl.BuildingName,
    COUNT(DISTINCT rl.RoomID) AS NumberOfRooms
FROM 
    [Location].[RoomLocation] rl
JOIN 
    [Location].[BuildingLocation] bl ON rl.BuildingCode = bl.BuildingID
GROUP BY 
    bl.BuildingName
ORDER BY 
    bl.BuildingName;
```

**Proposition 8**: Find all the Biochemistry Labs available for this semester.

```sql
USE QueensClassSchedule

SELECT c.CourseName,
       c.Instructor AS [Instructor Name],
       Enrolled AS [Enrollment Number],
       Limit AS [Enrollment Limit]
FROM CollegeClasses.Course as a
left outer join CollegeClasses.Class as c
on c.CourseID = a.CourseID
WHERE c.CourseName = 'Biochemistry Lab'
ORDER BY c.Enrolled;
```

**Propostion 9:** Determine the total number of courses per department

```sql
USE QueensClassSchedule

SELECT 
    CourseAbbreviation,
    COUNT(*) AS NumberOfCourses
FROM 
    [CollegeClasses].[Course]
GROUP BY 
    CourseAbbreviation
ORDER BY 
    CourseAbbreviation;
```

**Proposition 10:** Determine the course names that have less than 20 enrolled students.

```sql
USE QueensClassSchedule

SELECT 
    CourseName,
    Enrolled
FROM 
    [CollegeClasses].[Class]
WHERE 
    CAST(Enrolled AS INT) < 20
ORDER BY 
    CourseName;
```

**Proposition 11:** Determine all classes in the Science building "SB", along with their time.

```sql
USE QueensClassSchedule

SELECT 
    CourseAbbreviation,
    CourseNumber,
    CourseName,
    RoomNumber,
    [Time],
    [Day]
FROM 
    [CollegeClasses].[Class]
WHERE 
    BuildingName = 'SB'
ORDER BY 
    [Time];
```

**Proposition 12:** Determine the classes that are held on a Saturday or Sunday.

```sql
USE QueensClassSchedule

SELECT 
    CourseAbbreviation,
    CourseName,
    Instructor,
    BuildingName,
    RoomNumber,
    Day,
    [Time]
FROM 
    [CollegeClasses].[Class]
WHERE 
    [Day] IN ('S', 'Su')
ORDER BY 
    [Day], [Time];
```

**Proposition 13:** Determine classes that are hybrid and web-enhanced with a limit bigger than 50.

```sql
USE QueensClassSchedule

SELECT 
    CourseAbbreviation,
    CourseName,
    Instructor,
    BuildingName,
    RoomNumber,
    [Limit]
FROM 
    [CollegeClasses].[Class]
WHERE 
    ModeOfInstruction IN ('Hybrid', 'Web-Enhanced') AND 
    CAST([Limit] AS INT) > 50
ORDER BY 
    CourseAbbreviation, 
    CourseNumber;
```
